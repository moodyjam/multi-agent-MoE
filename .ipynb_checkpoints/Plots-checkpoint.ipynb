```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
data_to_plot = []
for file in os.listdir("data_to_plot"):
    path = os.path.join("data_to_plot", file)
    data = pd.read_csv(path)
    data_to_plot.append(data)
```

```python
os.listdir("data_to_plot")
```

```python
def plot(name_to_use, data_to_plot):
    # Adding labels and title.
    plt.xlabel('Step')
    plt.ylabel('Accuracy')
    plt.title(f'{name_to_use} Validation Accuracy')
    for df in data_to_plot:
        name = df.columns[1]
        if name_to_use in name:
            plt.figure(figsize=(10, 6))
            plt.plot(df['Step'], df[name], label=name)
    plt.legend()
        
        
```
