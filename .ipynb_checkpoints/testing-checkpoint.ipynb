```python
import networkx as nx
import random

def create_graph_with_connectivity(target_connectivity, num_nodes, max_iterations=1000):
    # Start with a path graph
    G = nx.path_graph(num_nodes)
    current_connectivity = nx.algebraic_connectivity(G)

    iteration = 0
    while abs(current_connectivity - target_connectivity) > 0.01 and iteration < max_iterations:
        # Modify the graph to adjust its connectivity
        # This could involve adding or removing edges, or changing the graph structure in other ways

        # Example modification: add a random edge to increase connectivity
        if current_connectivity < target_connectivity:
            a, b = sorted(random.sample(range(num_nodes), 2))
            if not G.has_edge(a, b):
                G.add_edge(a, b)
        # Example modification: remove a random edge to decrease connectivity
        else:
            edges = list(G.edges())
            if len(edges) > 1:
                G.remove_edge(*random.choice(edges))

        current_connectivity = nx.algebraic_connectivity(G)
        iteration += 1

    return G, current_connectivity
```

```python
# Usage
target_connectivity = 1.5  # Example target value
num_nodes = 10  # Example number of nodes
G, final_connectivity = create_graph_with_connectivity(target_connectivity, num_nodes)

print(f"Final algebraic connectivity: {final_connectivity}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

def create_graph_with_connectivity(target_connectivity, num_nodes, max_iterations=1000):
    G = nx.path_graph(num_nodes)
    current_connectivity = nx.algebraic_connectivity(G)

    iteration = 0
    while abs(current_connectivity - target_connectivity) > 0.01 and iteration < max_iterations:
        if current_connectivity < target_connectivity:
            a, b = sorted(random.sample(range(num_nodes), 2))
            if not G.has_edge(a, b):
                G.add_edge(a, b)
        else:
            edges = list(G.edges())
            if len(edges) > 1:
                G.remove_edge(*random.choice(edges))

        current_connectivity = nx.algebraic_connectivity(G)
        iteration += 1

    return G, current_connectivity

# Example usage
target_connectivity = 1.5
num_nodes = 10
G, final_connectivity = create_graph_with_connectivity(target_connectivity, num_nodes)

# Now, to visualize the graph:
plt.figure(figsize=(8, 6))  # Set the figure size (optional)
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray')
plt.title(f"Graph with Algebraic Connectivity: {final_connectivity:.2f}")
plt.show()
```
