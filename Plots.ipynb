```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
data_to_plot = []
for file in os.listdir("data_to_plot"):
    path = os.path.join("data_to_plot", file)
    data = pd.read_csv(path)
    data_to_plot.append(data)
```

```python
os.listdir("data_to_plot")
```

```python
def plot(name_to_use, data_to_plot):
    # Adding labels and title.
    plt.figure(figsize=(10, 6))
    plt.xlabel('Step')
    plt.ylabel('Accuracy')
    # plt.ylim(0,1)
    if name_to_use == "routing":
        name_to_use_full = "Routing Top 2"
    else:
        name_to_use_full = name_to_use.capitalize()
    plt.title(f'{name_to_use_full} Validation Accuracy')
    for df in data_to_plot:
        name = df.columns[1]
        if name_to_use in name:
            if name_to_use == "MNIST" and "FashionMNIST" in name:
                continue
            if "wobbly" in name:
                label = "DiNNO"
                color = "green"
            elif "oracle" in name:
                label = "MoE Oracle"
                color = "orange"
            else:
                label = "MoE"
                color = "blue"
            plt.plot(df['Step'], df[name], label=label.capitalize(), color=color)
    plt.legend()
```

```python
plot("FashionMNIST", data_to_plot)
```

```python
plot("CIFAR-10", data_to_plot)
```
